```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
numberOfNodes= 10
density= 0.3
l= int(np.sqrt(numberOfNodes/density))
Range= range(1, numberOfNodes)
```

```python
G = nx.Graph()
G.add_nodes_from(Range)
```

```python
pos = {}
i=1
while (i <=numberOfNodes):    
    val= pos.values()
    cordinate= (np.random.randint(0,l),np.random.randint(0,l))
    if cordinate not in val:
        pos.update({i:cordinate})
        i+=1
```

```python
for i in Range:
    for j in range(i+1, numberOfNodes):
        if (np.random.uniform()< 0.4):
            G.add_edge(i,j)
```

```python
#pos = {1: (0, 0), 2: (-1,-1), 3: (2,1), 4: (4, 0), 5: (5, 2)}
options = {
    "font_size": 13,
    "node_size": 300,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 3,
    "width": 3,
}
```

```python
nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.1)
plt.axis("off")
plt.show()
```
